```python
import requests

# Default
url = "https://tiki.vn/api/personalish/v1/blocks/listings"


# @ Change Referer:
# Remain: https://tiki.vn/
# Change: nha-sach-tiki & 8322 (Remain 'c')
# Example: "https://tiki.vn/thiet-bi-kts-phu-kien-so/c1815"

headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) Gecko/20100101 Firefox/147.0",
    "Accept": "application/json, text/plain, */*",
    "Referer": "https://tiki.vn/nha-sach-tiki/c8322"
}

# Change:
# - Category
# - Page
# - Sort
# - Urlkey
# - limit: default

params = {
    "category": 8322,
    "page": 1,
    "limit": 40,
    "sort": "top_seller",
    "urlKey": "nha-sach-tiki"
}

r = requests.get(url, headers=headers, params=params, timeout=30)
print(r.status_code)
print(r.json().keys())
```

```python
# Run to check the data
j = r.json()

type(j["data"]), len(j["data"]) if isinstance(j["data"], list) else j["data"].keys()
```

```python
type(j["items"]), len(j["items"]) if isinstance(j["items"], list) else j["items"].keys()
```

```python
# Check the features
if isinstance(j["data"], list) and len(j["data"]) > 0:
    sample = j["data"][0]
else:
    sample = j["items"][0]

sample.keys()
```

```python
from datetime import datetime
import pandas as pd

def normalize_quantity_sold(x):
    if isinstance(x, dict):
        return x.get("value")
    return x

products = j["data"]

rows = []
for p in products:
    rows.append({

        # ===== Identity =====
        "id": p.get("id"),
        "sku": p.get("sku"),
        "name": p.get("name"),
        
        # ===== Author / Brand =====
        # ===== Product grouping =====
        "primary_category_path": p.get("primary_category_path"),

        # ===== Pricing =====
        "price": p.get("price"),
        "original_price": p.get("original_price"),
        "discount": p.get("discount"),
        "discount_rate": p.get("discount_rate"),

        # ===== Trust / Social proof =====
        "rating_average": p.get("rating_average"),
        "review_count": p.get("review_count"),       
        # ===== Sales / Exposure =====
        "quantity_sold": normalize_quantity_sold(p.get("quantity_sold")),

        # ===== Inventory / Availability =====
        "availability": p.get("availability"),
        "shippable": p.get("shippable"),

        # ===== Seller =====
        "seller_id": p.get("seller_id"),
        "seller_product_id": p.get("seller_product_id"),

        # ===== Finance =====
        # ===== Full URL =====
        "url": "https://tiki.vn/" + (p.get("url_path") or ""),

    })

df = pd.DataFrame(rows)
df.head()
```

```python
import time
import requests

all_rows = []

def normalize_quantity_sold(x):
    if isinstance(x, dict):
        return x.get("value")
    return x

for page in range(1, 3):   # change 6 -> more pages if you want
    params["page"] = page

    r = requests.get(url, headers=headers, params=params, timeout=30)
    r.raise_for_status()
    j = r.json()

    products = j["data"]
    if not products:   # stop if empty
        break

    for p in products:
        all_rows.append({

        # ===== Identity =====
        "id": p.get("id"),
        "sku": p.get("sku"),
        "name": p.get("name"),
        
        # ===== Author / Brand =====
        # ===== Product grouping =====
        "primary_category_path": p.get("primary_category_path"),

        # ===== Pricing =====
        "price": p.get("price"),
        "original_price": p.get("original_price"),
        "discount": p.get("discount"),
        "discount_rate": p.get("discount_rate"),

        # ===== Trust / Social proof =====
        "rating_average": p.get("rating_average"),
        "review_count": p.get("review_count"),
       
        # ===== Sales / Exposure =====
        "quantity_sold": normalize_quantity_sold(p.get("quantity_sold")),

        # ===== Inventory / Availability =====
        "availability": p.get("availability"),
        "shippable": p.get("shippable"),

        # ===== Seller =====
        "seller_id": p.get("seller_id"),
        "seller_product_id": p.get("seller_product_id"),

        # ===== Finance =====
        # ===== Full URL =====
        "url": "https://tiki.vn/" + (p.get("url_path") or ""),
    })

    time.sleep(1)  # polite crawling

df_all = pd.DataFrame(all_rows)
df_all.shape, df_all.head()
```
