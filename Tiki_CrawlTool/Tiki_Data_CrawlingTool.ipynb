```python
import requests

# Default
url = "https://tiki.vn/api/personalish/v1/blocks/listings"


# @ Change Referer:
# Remain: https://tiki.vn/
# Change: nha-sach-tiki & 8322 (Remain 'c')
# Example: "https://tiki.vn/thiet-bi-kts-phu-kien-so/c1815"

headers = {
    "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) Gecko/20100101 Firefox/147.0",
    "Accept": "application/json, text/plain, */*",
    "Referer": "https://tiki.vn/the-thao-da-ngoai/c1975"
}

# Change:
# - Category
# - Page
# - Sort
# - Urlkey
# - limit: default

params = {
    "category": 1975,
    "page": 1,
    "limit": 40,
    "sort": "top_seller",
    "urlKey": "the-thao-da-ngoai"
}

r = requests.get(url, headers=headers, params=params, timeout=30)
print(r.status_code)
print(r.json().keys())
```

```python
# Run to check the data
j = r.json()

type(j["data"]), len(j["data"]) if isinstance(j["data"], list) else j["data"].keys()
```

```python
type(j["items"]), len(j["items"]) if isinstance(j["items"], list) else j["items"].keys()
```

```python
# Check the features
if isinstance(j["data"], list) and len(j["data"]) > 0:
    sample = j["data"][0]
else:
    sample = j["items"][0]

sample.keys()
```

```python
import pandas as pd

def normalize_quantity_sold(x):
    if isinstance(x, dict):
        return x.get("value")
    return x

products = j["data"]

rows = []
for p in products:
    rows.append({
        # ===== Identity =====
        "id": p.get("id"),
        "sku": p.get("sku"),
        "name": p.get("name"),
        "type": p.get("type"),
        "url_path": p.get("url_path"),

        # ===== Author / Brand =====
        "author_name": p.get("author_name"),
        "brand_name": p.get("brand_name"),

        # ===== Product grouping =====
        "productset_id": p.get("productset_id"),
        "productset_group_name": p.get("productset_group_name"),
        "primary_category_path": p.get("primary_category_path"),

        # ===== Pricing =====
        "price": p.get("price"),
        "list_price": p.get("list_price"),
        "original_price": p.get("original_price"),
        "discount": p.get("discount"),
        "discount_rate": p.get("discount_rate"),
        "bundle_deal": p.get("bundle_deal"),

        # ===== Trust / Social proof =====
        "rating_average": p.get("rating_average"),
        "review_count": p.get("review_count"),
        "order_count": p.get("order_count"),
        "favourite_count": p.get("favourite_count"),

        # ===== Badges =====
        "badges": p.get("badges"),
        "badges_new": p.get("badges_new"),
        "badges_v3": p.get("badges_v3"),

        # ===== Sales / Exposure =====
        "quantity_sold": normalize_quantity_sold(p.get("quantity_sold")),
        "product_reco_score": p.get("product_reco_score"),
        "advertisement": p.get("advertisement"),

        # ===== Inventory / Availability =====
        "inventory_status": p.get("inventory_status"),
        "inventory": p.get("inventory"),
        "stock_item": p.get("stock_item"),
        "salable_type": p.get("salable_type"),
        "availability": p.get("availability"),
        "shippable": p.get("shippable"),

        # ===== Seller =====
        "seller": p.get("seller"),
        "seller_id": p.get("seller_id"),
        "seller_product_id": p.get("seller_product_id"),

        # ===== Finance =====
        "is_gift_card": p.get("is_gift_card"),
        "installment_info": p.get("installment_info"),

        # ===== Full URL =====
        "url": "https://tiki.vn/" + (p.get("url_path") or "")
    })

df = pd.DataFrame(rows)
df.head()
```
