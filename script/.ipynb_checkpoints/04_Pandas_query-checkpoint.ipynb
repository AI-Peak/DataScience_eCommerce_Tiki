```python
import pandas as pd
from pathlib import Path

DATA_PATH = Path("..") / "datasets" / "cleaned_data" / "products_clean.csv"
df = pd.read_csv(DATA_PATH)

df.shape
```

```python
df.head()
df.describe()
```

Q1

```python
q1 = (
    df.groupby("category")
      .agg(
          n_products=("id", "count"),
          avg_price=("price", "mean"),
          median_price=("price", "median"),
          avg_rating=("rating_average", "mean"),
          avg_sold=("quantity_sold", "mean")
      )
      .sort_values("avg_sold", ascending=False)
)

q1
```

Q2

```python
#Uoc luong gmv = price x quantity_sold
df["gmv_proxy"] = df["price"] * df["quantity_sold"]

q2 = (
    df.groupby("category")["gmv_proxy"]
      .sum()
      .sort_values(ascending=False)
)

q2
```

```python
#tính cumulative % 
pareto = q2.cumsum() / q2.sum()
pareto
```

Q3 - Discount Strategy

```python
q3 = (
    df.groupby("category")
      .agg(
          avg_discount_rate=("discount_rate", "mean"),
          avg_discount=("discount", "mean")
      )
      .sort_values("avg_discount_rate", ascending=False)
)

q3
```

#### Q4 - Best seller Profile

```python
df_best = df[df["best_seller"] == True]
df_non = df[df["best_seller"] == False]

comparison = pd.DataFrame({
    "avg_price_best": df_best["price"].mean(),
    "avg_price_non": df_non["price"].mean(),
    "avg_rating_best": df_best["rating_average"].mean(),
    "avg_rating_non": df_non["rating_average"].mean(),
    "avg_discount_best": df_best["discount_rate"].mean(),
    "avg_discount_non": df_non["discount_rate"].mean(),
}, index=[0])

comparison
```

Q5 - Stock & Availability Impact

```python
q5 = (
    df.groupby("inventory_status")["quantity_sold"]
      .mean()
      .sort_values(ascending=False)
)

q5
```

Q6 — Trust Signals Impact

```python
trust_cols = [
    "rating_average",
    "reviews_count",
    "favourite_count"
]

df[trust_cols + ["quantity_sold"]].corr()["quantity_sold"]
```

Q7

```python
q7 = (
    df.groupby("age_bin", observed=True)
      .agg(
          avg_sold=("quantity_sold", "mean"),
          avg_rating=("rating_average", "mean")
      )
)

q7
```
